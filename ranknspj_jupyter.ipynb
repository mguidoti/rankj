# PLAZI: RANKING JOURNALS FOR TEMPLATE CREATION

## Top Journals for NEW SPECIES

```python
import pandas as pd

threshold = 51 # Defines the threshold to cut the dataframe and plot the chart

df = pd.read_csv('results\\nsp_out.csv', sep=';', engine='python') # Load the csv into a pd DataFrame

df.sort_values(by=['Proportion'], ascending=False) # Sort from higher to smaller

df_selected = df[df['Accumulated'] < threshold] # Selected results based on the defined threshold

print(df_selected[['Journal','Accumulated']]) # Print the selected journals
```

```python
# Plot a line plot based on the selected values
plot = df_selected.plot.line(x='Journal', y='Accumulated', figsize=(15,5))

# Make sure that all labels are being printed
plot.set_xticks(df_selected.index)
plot.set_xticklabels(df_selected.Journal, rotation=90);

# Add the 50% line
plot.axhline(y=50, color='red', linestyle='--')

```
